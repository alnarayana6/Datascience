```python
#Q.8) write a function to label an incident that was both successful and suicidal


su= (lambda x,y : x&y)
success=su(terror['suicide'], terror['success'])
```

```python
#Q.9) Create a new category representing if the incident occured in Afghanistan,
#Pakistan or India as one level of the category and all the other countries as
#another level. Label all other countries as ROW and new column which contains
#the new category as ‘Local’ in the data.
terror['Local']=np.where(terror['country_txt'].str.contains('India')|terror['country_txt'].str.contains('Pakistan')|terror['country_txt'].str.contains('Afghanistan'), "Local","ROW")
```

```python
#Q.10) How many incidents happened in Af-Pak-India vs ROW?

terror['Local'].value_counts()
```

```python
#Q.11) List the number of suicides attacks and average kills by Af-Pak-India vsROW. Rename columns in the output as Average_Kills for average kills and
#Number_Incidents for count of suicide attacks.


terror.groupby(["Local","suicide"],as_index=False).agg({'nkill':np.mean}).rename(columns={"nkill":"Average_kills"})
```

```python
terror.groupby(["Local","suicide"],as_index=False).agg({'nkill':np.count_nonzero}).rename(columns={"nkill":"Number_Incidents"})
```
